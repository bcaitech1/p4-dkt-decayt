```python
import os

import numpy as np
import pandas as pd
```

```python
path_dir = './output/ensemble/'
 
file_list = os.listdir(path_dir)

# ensemble_list = [pd.read_csv(path_dir + f)for f in file_list]
ensemble_list = [
pd.read_csv(path_dir+'output (12).csv'),
pd.read_csv(path_dir+'output (13).csv'),
pd.read_csv(path_dir+'output (14).csv'),
pd.read_csv(path_dir+'output (15).csv'),
pd.read_csv(path_dir+'output (16).csv'),
]
```

```python
df = ensemble_list[0].copy()
for i in range(1, len(ensemble_list)):
    df['prediction'] += ensemble_list[i]['prediction']

df['prediction'] = df['prediction']/len(ensemble_list)
```

```python
print(df[df['prediction'] >= 0.5].count())
print(ensemble_list[1].loc[ensemble_list[1].prediction >= 0.5].count())
```

```python
df.to_csv('ensemble.csv', index=False)
```
